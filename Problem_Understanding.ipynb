# Customer Churn Prediction Case Study

## Problem Understanding

Customer churn is one of the biggest challenges for subscription-based businesses such as telecom companies, streaming platforms, and banks.

When customers stop using a company’s services, the company loses revenue and may also spend additional money to acquire new customers.

The main business problem is:

"How can the company identify customers who are likely to leave?"

By predicting churn early, the company can take preventive actions such as:
- providing special offers,
- improving customer support,
- creating loyalty programs,
- and increasing customer satisfaction.

## Project Goal

The goal of this project is to build a machine learning model that predicts whether a customer is likely to churn based on customer data such as:
- contract type,
- monthly charges,
- tenure,
- payment method,
- and internet services.

The insights generated from this analysis can help businesses reduce churn and improve customer retention.
